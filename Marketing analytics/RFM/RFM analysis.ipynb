### Об RFM-анализе

#### Введение

RFM-анализ (от англ. *Recency, Frequency, Monetary*) — это метод сегментации и оценки клиентской базы компании, который позволяет более эффективно управлять маркетингом и взаимодействием с клиентами. Данный метод был впервые представлен в 1995 году в статье Я.Р. Бальта (Bult) и Т. Вансбика (Wansbeek) «Оптимальный выбор для прямой почтовой рассылки», где авторы подтвердили принцип Паретто —  идею о том, что 80% продаж приходится на 20% клиентов бренда.

Метод основан на анализе комбинаций из трех ключевых параметров:

1. **Recency** (давность последней покупки) — сколько времени прошло с момента последней покупки клиента.
2. **Frequency** (частота покупок) — сколько раз клиент совершал покупки за определенный период.
3. **Monetary** (сумма потраченных средств) — общая сумма, потраченная клиентом за этот период.

Этот анализ позволяет сегментировать клиентов на основе их покупательской активности, что помогает компаниям оптимизировать свои маркетинговые стратегии и ресурсы.

#### Цели RFM-анализа

Основные цели RFM-анализа включают:
1. **Сегментация клиентов** — разделение клиентской базы на группы по их активности, что помогает сосредоточить усилия на наиболее перспективных сегментах.
2. **Повышение эффективности маркетинговых кампаний** — компания может разрабатывать персонализированные предложения для разных категорий клиентов.
3. **Прогнозирование поведения** — анализ позволяет предсказать будущие действия клиентов на основе их предыдущей активности.
4. **Удержание клиентов** — помогает определить клиентов, которые могут перестать взаимодействовать с брендом, и принять меры для их удержания.
5. **Повышение доходов** — с помощью анализа компании могут точнее понимать, какие клиенты приносят больше прибыли и как стимулировать их к дальнейшим покупкам.

#### Принципы RFM-анализа

RFM-анализ основывается на трех показателях, каждый из которых играет важную роль в понимании клиентского поведения:

1. **Recency (давность последней покупки)**:
   - Чем меньше времени прошло с момента последней покупки, тем вероятнее, что клиент снова совершит покупку.
   - Покупатели, которые недавно взаимодействовали с компанией, как правило, более лояльны и активны.

2. **Frequency (частота покупок)**:
   - Часто покупающие клиенты обычно проявляют высокий уровень лояльности к бренду.
   - Чем чаще клиент покупает, тем более выгодно его удержание для компании.

3. **Monetary (сумма покупок)**:
   - Определяет, насколько ценен клиент с точки зрения финансов.
   - Те, кто тратят больше денег, как правило, обладают большей значимостью для бизнеса.

#### Методология проведения RFM-анализа

Для проведения RFM-анализа необходимо собрать данные по каждому клиенту о его покупках и поведенческих метриках. Обычно процесс состоит из нескольких этапов:

1. **Сбор данных**:
   - Необходимо собрать данные о покупках каждого клиента, включая дату последней покупки, количество совершенных покупок и общую сумму.

2. **Оценка клиентов**:
   - Каждому клиенту присваивается оценка по трем показателям (*Recency*, *Frequency* и *Monetary*). Например, клиент, недавно совершивший покупку, получает более высокий балл по параметру *Recency*.

3. **Сегментация**:
   - Клиенты делятся на группы на основе полученных оценок. Например, могут быть выделены такие группы, как «наиболее лояльные клиенты», «недавние покупатели» и «клиенты с высокой частотой покупок, но низкими тратами».

4. **Интерпретация данных**:
   - На основе сегментации можно делать выводы о стратегиях взаимодействия с клиентами. Например, для группы «давно неактивных клиентов» можно запустить программу стимулирования к покупкам через персонализированные предложения.

#### Применение RFM-анализа

RFM-анализ широко применяется в различных отраслях, таких как ритейл, банковское дело, телекоммуникации и электронная коммерция. Примеры применения:

1. **Персонализированный маркетинг**:
   - Компании могут создавать уникальные предложения для клиентов с высоким значением *Frequency*, чтобы укрепить их лояльность и побудить к новым покупкам.

2. **Оптимизация рекламных кампаний**:
   - Клиенты с высокой оценкой *Recency* и *Frequency* более склонны реагировать на новые предложения, поэтому рекламные бюджеты могут быть распределены эффективнее.

3. **Анализ жизненного цикла клиента**:
   - RFM-анализ помогает определить стадию жизненного цикла каждого клиента и соответственно подбирать стратегию взаимодействия.

4. **Управление скидками**:
   - На основе анализа можно идентифицировать клиентов, которым нужно предложить скидки или бонусы для их возвращения в категорию активных покупателей.

#### Преимущества и недостатки RFM-анализа

**Преимущества**:
- Простота и доступность в использовании.
- Высокая эффективность при анализе больших клиентских баз (более 10 000 человек).
- Помогает выявить наиболее ценных клиентов.
- Возможность прогнозировать поведение клиентов на основе их покупательской истории.

**Недостатки**:
- Ограниченная точность, так как не учитывает более глубокие поведенческие аспекты (например, предпочтения и удовлетворенность).
- Может не учитывать сезонные колебания покупательской активности.
- Не всегда применим для новых клиентов, у которых нет обширной истории покупок.

Большинство полей этих таблиц не пригодятся нам в ходе анализа, поэтому в целях эконмии вычислительных мощностей возьмем только нужные нам данные и проведем предварительный разведочный анализ. Начнем таблицы пользователей

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
customers_df = pd.read_csv('olist_customers_dataset.csv')
customers_df.head()
```

Самыми важными для нас здесь являются поля `customer_unique_id` и `customer_id`, так как они потребуются для идентификации пользователя и его связи с заказами, которые он делал. Оставим только эти поля. Также убедимся, что эти поля не содержат пропусков:

```python
customers_df = customers_df[['customer_id', 'customer_unique_id']]
customers_df.info()
```

Теперь обратимся к таблице заказов.

```python
df_orders = pd.read_csv('olist_orders_dataset.csv')
df_orders.head()
```

Здесь оставим только поля `order_id`,	`customer_id`,	`order_status` `order_purchase_timestamp`. Также проверим полноту данных.

```python
df_orders = df_orders[['order_id', 'customer_id', 'order_status', 'order_purchase_timestamp']]
df_orders.info()
```

Из поля `order_purchase_timestamp` создадим новый признак - order_month к типу данных 'period[M]'

```python
df_orders['order_date'] = pd.to_datetime(df_orders['order_purchase_timestamp']).dt.date
df_orders.head()
```

Посмотрим на существующие статусы заказов:

```python
df_orders['order_status'].unique()
```

Заказы со статусами 'unavailable' и 'canceled' мы учитывать не будем.

```python
df_orders = df_orders.query('order_status != "unavailable" and order_status != "canceled"')[['order_id', 'customer_id', 'order_date']]
df_orders.head()
```

Перейдем к таблице товаров

```python
df_items = pd.read_csv('olist_order_items_dataset.csv')
df_items.head()
```

```python
# пропусков в данных не наблюдается:
df_items.info()
```

Перемножим стоимость каждого товара на его количество в заказе и предагрегируем данные на уровне заказа по полю `order_id`

```python
df_items['total_price'] = df_items['price'] * df_items['order_item_id']
```

```python
df_items = df_items.groupby('order_id').agg({'total_price': 'sum'})
df_items.head()
```

Данные подготовлены, можно выполнять слияние таблиц:

```python
df = df_orders \
        .merge(customers_df, on='customer_id', how='left')  \
        .merge(df_items, on='order_id', how='left')[['customer_unique_id', 'order_date', 'total_price']]
df.head()
```

Будем считать "сегодняшней" датой в рамках RFM-анализа дату последнего заказа в получившемся датафрейме. Исходя из этого, рассчитаем новый признак - количество дней, прошедших между датой заказа и днем сегодняшним.

```python
df['days_delta'] = (df['order_date'].max() - df['order_date']).apply(lambda x: x.days)
df.head()
```

Выберем временной период, на котором мы будем проводить RFM анализ. Чаще всего для этого берется период длительностью один год от текужей даты, но могут быть выбраныи и другие периоды, в зависимости от отрасли и вида бизнеса. Отфильтруем колонку `total_price` с учетом выбранного периода и далее выполним расчеты, необходимые для RFM-анализа:

```python
# период 1 год
period = 365

# не будем учитывать стоимости покупок вне перода анализа
df['total_price'][df['days_delta'] > period] = 0
```

```python
aggregations = {
    'days_delta': [lambda i: min([j if j <= period else period + 1 for j in i ]), # считаем сколько дней прошло с самого последнего заказа, при этом, если прошло более, чем period дней, то считаем, что прошло period дней
                   lambda x: len([y for y in x if y <= period])], # считаем количество заказов, при этом учитываем только заказы, для которых days_delta не более, чем period
    'total_price': 'sum'
}
```

```python
df = df.groupby('customer_unique_id') \
      .agg(aggregations) \
      .droplevel(0, axis=1) \
      .rename(columns={'<lambda_1>': 'orders_count', '<lambda_0>': 'days_delta', 'sum': 'orders_total_sum'}) \
      .reset_index()
df.head()
```

Теперь нам необходимо отнести каждого пользователя в некоторую группу по  соответствующему ему значению в каждом из полей таблицы.

При выборе стратегии для определения количества групп и их границ в RFM-анализе, важно учитывать несколько факторов: цель анализа, размер клиентской базы и детализацию, которую хотите получить. Существует много подходов и стратегий выбора оптимального количества групп и определения границы для каждой из них, например:

* Классификация с равными интервалами (*Equal Interval Binning*).
Этот метод заключается в делении каждого из показателей (Recency, Frequency и Monetary) на несколько равных интервалов. Это самый простой подход, при котором количество групп (классов) одинаково для всех показателей, например:

 - Разделить на 3 или 5 классов: высокий, средний, низкий (или более детализированные уровни, такие как «очень высокий», «очень низкий»).
* Метод перцентилей.
Этот метод заключается в разделении клиентов на группы, используя определенные перцентили. Например, можно разделить клиентов на группы по 25, 50 и 75 перцентилям для каждой метрики RFM. Это позволяет более гибко настраивать количество групп и границы.
* Классификация на основе кластерного анализа (*K-means Clustering*).
Этот метод использует алгоритмы кластеризации (например, *K-means*), чтобы автоматически находить группы клиентов на основе их поведения по трем показателям RFM. Модель машинного обоучения сама решает, как лучше сгруппировать клиентов, минимизируя внутригрупповые различия и максимизируя различия между группами. Данный метод подходит для больших наборов данных.
* Классификация на основе бизнес-логики.
Этот подход заключается в создании групп, исходя из бизнес-целей и понимания важности различных сегментов. Например, можно определить группы в зависимости от того, насколько клиент ценен для бизнеса.

В нашем случае для примера попробуем разделить клиентов на 3 группы:

| Давность последней покупки (Recency) | Частота покупок (Frequency) | Сумма покупки (Monetary)  |
|--------------------------------------|-----------------------------|----------------------------|
| 1 - недавние клиенты                 | 1 - частые                  | 1 - большая                |
| 2 - редкие нерегулярные              | 2 - редкие                  | 2 - средняя                |
| 3 - давние, возможно потерянные      | 3 - очень редкие            | 3 - маленькая              |

Для того, чтобы определить границы между группам, посмотрим на каждый признак в отдельности.

### Recency (давность последней покупки):

```python
df.days_delta.value_counts(bins=10)
```

Здесь можно разбить на 3 группы по количеству дней без покупок: 0 - 90, 90 - 270, 270+

```python
df['R'] = df.days_delta.apply(lambda x: 1 if x <= 90 else (2 if x <= 270 else 3))
```

### **Frequency** (частота покупок)

```python
df.orders_count.value_counts()
```

По количеству заказов можно выделить 3 группы с границами [0 - 1), [1, 2), [2, 11]

```python
df['F'] = df.orders_count.apply(lambda x: 3 if x < 1 else (2 if x < 2 else 1))
```

### **Monetary** (сумма потраченных средств)

Здесь мы разобъем на 3 группы согласно принципу Паретто 80/20:

```python
df['M'] = pd.qcut(df.orders_total_sum, [0, 0.8, 0.96, 1], labels=[3, 2, 1])
```

### Посмотрим на получившийся результат

```python
df.sort_values(by=['R', 'F', 'M'], ascending=True)
```

| **R2 - Средней давности клиенты** | Некоторое время не покупали                 | - Запустите кампании повторного привлечения, используя скидки или эксклюзивные предложения.<br> - Напомните об оставленных товарах в корзине или просмотренных ранее продуктах.            |

#### 2. Стратегии для сегментов по Frequency (частота покупок)

|-----------------------------------|---------------------------------------------|--------------------------------------------------------------------------------------------------------|

#### 3. Стратегии для сегментов по Monetary (сумма покупок)

| **Группа**                        | **Характеристика**                          | **Стратегия**                                                                                          |
|-----------------------------------|---------------------------------------------|--------------------------------------------------------------------------------------------------------|
| **M1 - Высокие суммы покупок**    | Клиенты, тратящие много                     | - Предлагайте эксклюзивные премиальные товары или VIP-обслуживание.<br> - Рассмотрите персонализированные предложения, доступ к новым продуктам до их официального запуска. |
| **M2 - Средние суммы покупок**    | Клиенты со средней корзиной                 | - Подталкивайте к покупкам большего объема через скидки при достижении определенной суммы заказа.<br> - Организуйте акции типа «купи больше – получи больше» для увеличения среднего чека. |
| **M3 - Низкие суммы покупок**     | Клиенты, тратящие мало                      | - Создайте кампании для стимулирования небольших покупок, например, «товары до $X» или «недорогие подарки».<br> - Предложите скидки на связки товаров для увеличения суммы заказа. |

#### Комбинированные стратегии для сегментов RFM

В RFM-анализе клиенты попадают в одну из групп по каждому из показателей, что создает множество комбинаций. Например, клиент может быть **R1F1M1** — частый и недавний покупатель, который тратит много, или **R3F3M3** — клиент, давно не покупающий, редко совершавший покупки с небольшими суммами.

#### Примеры стратегий для комбинированных групп:

1. **R1F1M1 (активные и ценные клиенты)**:
   - Поддерживайте высокий уровень лояльности с помощью эксклюзивных предложений, программ лояльности и персонализированных скидок.
   - Вовлекайте их в реферальные программы, чтобы привлекать новых клиентов через рекомендации.
   - Устраивайте розыгрыши или привилегии за достижение определенных порогов покупок.
   - Повышенный кэшбек.


2. **R3F3M3 (потерянные клиенты с низкой активностью)**:
   - Реактивация через привлекательные предложения, такие как большие скидки или купоны на бесплатную доставку.
   - Отправка персонализированных сообщений с предложением восстановить отношения с брендом и напоминанием о преимуществах покупок у вас.
   - Альтернативная стратегия - отпустить и не тратить на этих клиентов ресурсы.
   

3. **R1F3M2 (недавние, но редко покупающие клиенты)**:
   - Стимулируйте их к частым покупкам через бонусные программы, где можно накапливать баллы за каждую покупку.
   - Предлагайте бесплатную доставку при повторных покупках в течение определенного срока (например, в течение 30 дней после первой покупки).


4. **R2F2M2 (среднеактивные клиенты со средней корзиной)**:
   - Привлекайте их к увеличению суммы заказов с помощью акций «купи на X, получи скидку Y» или предложений с бесплатной доставкой.
   - Поддерживайте интерес с помощью сезонных распродаж или эксклюзивных предложений по продуктам, которые могут их заинтересовать.




В заключение посмотрим, как распределились покупатели  в разрезе объемов продаж в зависимости от их RF группы:

```python
# plot the distribution of M for RF score
fig, axes = plt.subplots(nrows=3, ncols=3,
                         sharex=False, sharey=True,
                         figsize=(6, 6))

r_range = range(1, 4)
f_range = range(1, 4)
for r in r_range:
    for f in f_range:
        y = df[(df['R'] == r) & (df['F'] == f)]['M'].value_counts().sort_index()
        x = y.index
        ax = axes[r - 1, f - 1]
        bars = ax.bar(x, y, color='silver')
        if r == 3:
            if f == 2:
                ax.set_xlabel('{}\nF'.format(f), va='top')
            else:
                ax.set_xlabel('{}\n'.format(f), va='top')
        if f == 1:
            if r == 2:
                ax.set_ylabel('R\n{}'.format(r))
            else:
                ax.set_ylabel(r)
        ax.set_frame_on(False)
        ax.tick_params(left=False, labelleft=False, bottom=False)
        ax.set_xticks(x)
        ax.set_xticklabels(x, fontsize=8)

        for bar in bars:
            value = bar.get_height()
            if value == y.max():
                bar.set_color('firebrick')
            ax.text(bar.get_x() + bar.get_width() / 2,
                    value,
                    int(value),
                    ha='center',
                    va='bottom',
                    color='k')
fig.suptitle('Distribution of Monitaryu groups for each frequency and recency group',
             fontsize=14)
plt.tight_layout()
plt.show()
```
